```python
import pandas as pd
data = pd.read_csv("bank-additional-full.csv", sep=';')
```

```python
data.head()
```

```python
y = data['y']
data = data.drop('y', axis=1)
```

```python
y = y.astype('category')
y = y.cat.codes
```

Предварительная обработка данных:

```python
data.columns
```

```python
data.columns[data.isna().any()].tolist()
```

One-hot кодирование для необходимых признаков:

```python
names = ['job',	'marital',	'education', 'default',	'housing',	'loan',	'contact',	'month',	'day_of_week', 'poutcome']
for n in names:
  data[n] = data[n].astype('category')
  data[n] = data[n].cat.codes
  data = pd.get_dummies(data, columns=[n])
```

Рассмотрим 2 варианта: данные после нормализации и без неё

```python
data_max = data.max()
data_min = data.min()
data_norm = data
data_norm = data_norm - data_min
data_norm = data_norm.div(data_max - data_min)
data_norm.describe()
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, y_train, y_test = train_test_split(data, y, test_size=0.3)
data_train_norm, data_test_norm, y_train_norm, y_test_norm = train_test_split(data_norm, y, test_size=0.3)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
```

Обучение на данных без нормализации:

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(data_train, y_train)
predicted = model.predict(data_test)   # для вычисления accuracy
predicted_probs = model.predict_proba(data_test)[:, 1]   # для вычисления log_loss
```

```python
acc = accuracy_score(predicted, y_test)  # вычисление accuracy: на вход подаётся результат функции predict
loss = log_loss(y_test, predicted) # вычисление log_loss: на вход подаётся результат функции predict_proba

print("accuracy на тестовых данных: ", acc)
print("log_loss на тестовых данных: ", loss)
```

Обучение на нормализованных данных:

```python
model = LogisticRegression()
model.fit(data_train_norm, y_train_norm)
predicted = model.predict(data_test_norm)   # для вычисления accuracy
predicted_probs = model.predict_proba(data_test_norm)[:, 1]   # для вычисления log_loss
```

```python
acc = accuracy_score(predicted, y_test_norm)  # вычисление accuracy: на вход подаётся результат функции predict
loss = log_loss(y_test_norm, predicted) # вычисление log_loss: на вход подаётся результат функции predict_proba

print("accuracy на тестовых данных: ", acc)
print("log_loss на тестовых данных: ", loss)
```

Вычислим порог обзвона (вероятность успеха операции для конкретного клиента) и удельную прибыль компании за одного клиента из базы данных при наилучшем пороге обзвона, если каждый звонок стоит 10 руб., а успешный звонок приносит компании 100 руб:

```python
model = LogisticRegression()
model.fit(data_train, y_train)

predicted = model.predict(data_train)   # для вычисления accuracy
predicted_probs = model.predict_proba(data_train)[:, 1]   # для вычисления log_loss
```

```python
res = []
arrp = []
for t in range(0, 100):
  p = t / 100
  arrp += [p]
  waste = sum(predicted_probs >= p) * 10
  profit = 0
  for i in range(len(predicted_probs)):
    profit += int(predicted_probs[i] >= p and y_train.iloc[i] == 1) * 100
  res += [profit - waste]
```
